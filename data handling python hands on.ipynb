```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creating the dataset
data = {'Cricket':50, 'Basketball':15, 'Kabaddi':30,
        'Hockey':35, 'Football':10}
sports = list(data.keys())
values = list(data.values())
  
#fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(sports, values,
        width = 0.4)
 
plt.xlabel("Sport type")
plt.ylabel("No. of fans")
plt.title("Students favourite sport")
plt.show()
```

```python
X = ['Cricket','Basketball','Kabaddi','Hockey', 'Football']
girls = [10,20,20,40,20]
boys = [50,30,25,30,30]
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, girls, 0.4, label = 'Girls')
plt.bar(X_axis + 0.2, boys, 0.4, label = 'Boys')
  
plt.xticks(X_axis, X)
plt.xlabel("Groups")
plt.ylabel("Number of fans")
plt.title("Number of Students in each group")
plt.legend()
plt.show()
```

```python
# Data to plot
labels = 'Cricket','Basketball','Kabaddi','Hockey', 'Football'
sizes = [250, 100, 125, 210,25]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'gray']
explode = (0.1, 0, 0, 0,0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```
